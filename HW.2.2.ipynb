```python
# Порахуйте, яка частка пропусків міститься в кожній колонці

missing = df.isnull().sum()
total = len(df)
percentage = (missing / total) * 100
print(percentage)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

df = df.dropna(
    subset=[col for col in df.columns if col != 'Язык.программирования'])
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

missing = df.isnull().sum()
total = len(df)
percentage = (missing / total) * 100
print(percentage)
```

```python
# Видаліть усі рядки у вихідній таблиці

df = df.dropna()
```

```python
# Визначте новий розмір таблиці

df.shape
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

new_df = pd.read_csv("2017_jun_final.csv")

python_data = new_df.loc[df['Язык.программирования'] == 'Python']

print(python_data)
```

```python
# Визначте розмір таблиці python_data

python_data.dtypes
```

```python
# Визначте розмір таблиці python_data

python_data.shape
```

```python
# Bиконайте групування за стовпчиком "Посада"

grouped_data=python_data.groupby(by = 'Должность')
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

aggregated_data = grouped_data.agg({'': ['min', 'max']})

print(aggregated_data)
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

def fill_avg_salary(row):
    return row['Зарплата.в.месяц'].mean()


aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
print(aggregated_data)
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
avg_column_description = aggregated_data['avg'].describe()

print(avg_column_description)
```

```python
# Збережіть отриману таблицю в CSV файл

aggregated_data.to_csv("python_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#графік 1

new_data = df.groupby(by='Должность').agg({'Зарплата.в.месяц':'mean'})

plt.bar(new_data.index, new_data['Зарплата.в.месяц'], color = 'tomato')
plt.title('Залежність зарплати від посади',
          fontsize=15, color='darkred',)
plt.xlabel('Посада', fontsize='medium', color='firebrick')
plt.ylabel('Зарплата, $', fontsize='medium', color='firebrick')
plt.xticks(rotation=90)
```

```python
# графік 2

new_data = df['Пол'].value_counts().reset_index()

plt.pie(new_data['count'], labels=new_data['Пол'],
        autopct="%.2f%%", colors=['#005ce2', '#ff83d3'])

plt.show()
print(new_data)
```

```python
# графік 3

custom_colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']
sns.barplot(x="Уровень.английского", y="Зарплата.в.месяц",
            data=df, palette=custom_colors)
plt.xticks(rotation=45)
plt.xlabel('Рівень англійської', fontsize='medium', color='darkgoldenrod')
plt.ylabel('Зарплата на місяць, $', fontsize='medium', color='darkgoldenrod')
plt.title('Залежність зарплати від рівня англійської',
          fontsize=15, color='orchid',)
```

```python
# графік 4

new_data = df['Университет'].value_counts().reset_index()

plt.figure(figsize=(10, 8))
sns.barplot(x='count', y='Университет', data=new_data, palette='viridis')
plt.title('Кількість програмістів за університетами')
plt.xlabel('Кількість програмістів')
plt.ylabel('Університет')
plt.tight_layout()
plt.show()
```

```python
new_data = df.groupby(by="Общий.опыт.работы").agg(
    {'Зарплата.в.месяц': 'mean'}).reindex()

new_data = new_data.replace(
    to_replace='меньше 3 месяцев', value='меньше \n 3 \n месяцев)')

new_index = ['меньше 3 месяцев', "0.25", "0.5", "1", "1.5",
             "2", "3", "4", "5", "6", "7", "8", "9", "10 и более лет"]

new_data = new_data.reindex(new_index)

plt.figure(figsize=(10, 6))

sns.lineplot(x="Общий.опыт.работы", y="Зарплата.в.месяц",
             data=new_data, color='#7FFFD4')

plt.xticks(rotation=60)

plt.title('Залежність зарплати на місяць від стажу', fontsize='15')

plt.xlabel('Досвід роботи')
plt.ylabel('Зарплата на місяць')

print(new_data)
```
